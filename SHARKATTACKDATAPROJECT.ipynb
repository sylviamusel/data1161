### 4.3 Adding month column and making it into intergers

```python
sad['month'] = sad['Case Number'].str[5:7]
sad['month'] = sad['month'].astype('int')
import calendar 
def monthtostring(x):
    try:
        return calendar.month_name[x]
    except:
        return "unknown"
    

sad['month_name'] = sad.month.apply(monthtostring)

sad.head()
```

### 4.4 Checking for Nan

```python
sad3 = sad[sad['month'].isna()]
print(sad3)

sad3.head()
```

```python
results= sad.groupby('month').sum()
```

```python
sad[['First name','Last name']] = sad['Name'].loc[sad['Name'].str.split().str.len() == 2].str.split(expand=True)

#sad['First name'] = sad['First name'].replace(to_replace='Mr.', value='None')
sad['First name'] = sad['First name'].replace('Mr.', np.NaN)
sad['First name'] = sad['First name'].replace('2', np.NaN)
sad['First name'] = sad['First name'].replace('a', np.NaN)

sad.head()
```

```python
def tabless():

    from IPython.display import display_html
    from itertools import chain,cycle
    def display_side_by_side(*args,titles=cycle([''])):
        html_str=''
        for df,title in zip(args, chain(titles,cycle(['</br>'])) ):
            html_str+='<th style="text-align:center"><td style="vertical-align:top">'
            html_str+=f'<h2 style="text-align: center;">{title}</h2>'
            html_str+=df.to_html().replace('table','table style="display:inline"')
            html_str+='</td></th>'
        display_html(html_str,raw=True)

    grouped1= filtered_sad[['First name', 'Activity', 'Country']].value_counts().reset_index(name='# of attacks')
    grouped1=grouped1.head(20)

    grouped2= filte2_sad[['First name', 'Activity', 'Country']].value_counts().reset_index(name='# of attacks')
    grouped2=grouped2.head(20)

    display_side_by_side(grouped1, grouped2, titles=['1940-1980','1980-2020'])
```
